```python
weights_list = []
names_list = []
```

```python
for fn_comp in glob.glob("/Users/Trevor1/Dropbox/HST/c3_priors/MattS_CIV_tplte/*.fits"):
    comp = fits.open(fn_comp)
    names_list.append(fn_comp.split("/")[-1])
    name = fn_comp.split("/")[-1][:-5] ; name_title = "-".join(name.split("_"))
    c0 = comp[0].header["COEFF0"]
    c1 = comp[0].header["COEFF1"]
    npix = comp[0].header["NGOOD"]
    wave = 10.**np.linspace(c0, c0+c1*(npix-1), npix)
    flux = comp[0].data[0,:]
    errs = comp[0].data[1,:]
    
    wave_arb, flux_arb, errs_arb, mask_arb, wave_ica, flux_ica, f2500_ica, weights = run_ICA_r20_components_priors.main_ICA(wave, flux, errs, 
                                                                          np.zeros(len(wave)), z=0, name="", 
                                        ica_path="/Users/Trevor1/Dropbox/ICA_module/components/")
    
    weights_list.append(weights)
    fig, ax = plt.subplots(1, 1, figsize=(15,5))
    #ax.plot(wave, flux, "-k", lw=4)
    ax.plot(wave_arb, flux_arb, "-k", lw=4)
    ax.plot(wave_ica, flux_ica, "-r")
    #ax.plot(wave, errs, "-k", lw=2, alpha=0.3)
    
    ax.set_xscale("log")
    ax.set_yscale("log")
    ticks = np.array([1200,1500,1750,2000,2200,2500,2800])
    ax.set_xticks(ticks)
    ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
    ax.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
    ax.get_xaxis().set_tick_params(which='minor', size=0)
    ax.get_xaxis().set_tick_params(which='minor', width=0)
    plt.minorticks_off()
    ax.set_xlim(min(wave), max(wave))
    ax.set_title(name_title+" mod EW fit")
    plt.tight_layout()
    #plt.savefig("composites_pdf/modEWcomp_fits/%s.pdf"%name_title)
    plt.show()
    #break
```

```python
df_weights = pd.DataFrame(np.array(weights_list), columns=["W1","W2","W3","W4","W5","W6","W7","W8","W9","W10",])
df_weights.insert(loc=0, column='Composite_Name', value=names_list)
df_weights
```
